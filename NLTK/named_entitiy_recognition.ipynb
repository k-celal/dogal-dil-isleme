```python
import nltk
```

```python
text="William Henry Gates III (born October 28, 1955) is an American businessman, investor, philanthropist, computer programmer, and writer best known for co-founding the software giant Microsoft, along with his childhood friend Paul Allen.[2][3] During his career at Microsoft, Gates held the positions of chairman, chief executive officer (CEO), president, and chief software architect, while also being its largest individual shareholder until May 2014.[4] He was a major entrepreneur of the microcomputer revolution of the 1970s and 1980s."
```

```python
tokenized = nltk.word_tokenize(text)
```

```python
tagged=nltk.pos_tag(tokenized)
```

```python
ner=nltk.ne_chunk(tagged)
```

```python
ner.draw()
```
